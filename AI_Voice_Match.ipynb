```
option = "Upload Your Own Voice" #@param ["Default Voice For Testing", "Upload Your Own Voice"]
#filename = "elon audio.wav" #@param {type: "string"}
if option=="Default Voice For Testing":
  in_fpath = Path("pretrained_models/elon audio.wav")
else:
  from google.colab import files
  uploaded = files.upload()
  filepath = next(iter(uploaded))
  in_fpath = Path(filepath)
clear_output()
reprocessed_wav = encoder.preprocess_wav(in_fpath)
original_wav, sampling_rate = librosa.load(in_fpath)
preprocessed_wav = encoder.preprocess_wav(original_wav, sampling_rate)
embed = encoder.embed_utterance(preprocessed_wav)
text = "Type the text that you want the AI to speak" #@param {type:"string"}
with io.capture_output() as captured:
  specs = synthesizer.synthesize_spectrograms([text], [embed])
generated_wav = vocoder.infer_waveform(specs[0])
generated_wav = np.pad(generated_wav, (0, synthesizer.sample_rate), mode="constant")
from IPython.display import clear_output
clear_output()
display(Audio(generated_wav, rate=synthesizer.sample_rate))
```
